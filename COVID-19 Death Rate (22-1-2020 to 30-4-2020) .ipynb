```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
corona_deaths = pd.read_csv('Downloads\covid19_deaths_dataset.csv')
```

```python
corona_deaths.head(10)
```

```python
corona_deaths.drop(['Lat','Long'], axis=1, inplace=True)
```

```python
covid19_deaths = corona_deaths.groupby('Country/Region').sum()
```

```python
covid19_deaths.head()
```

```python
covid19_deaths.shape
```

```python
max_death_rate = []
countries = covid19_deaths.index
for country in countries:
    max_death_rate.append(covid19_deaths.loc[country].diff().max())
max_death_rate
```

```python
covid19_deaths['max death rate'] = max_death_rate
```

```python
covid19_deaths.head()
```

```python
covid19_deaths = pd.DataFrame(covid19_deaths.loc[:, 'max death rate'])
```

```python
covid19_deaths.shape
```

```python
covid19_deaths.to_csv('Downloads\covid19_deaths.csv')
```
