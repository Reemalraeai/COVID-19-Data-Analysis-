```python
corona_dataset.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
corona_dataset.head()
```

```python
corona_dataset2 = corona_dataset.groupby('Country/Region').sum()
```

```python
corona_dataset2.head()
```

```python
corona_dataset2.loc['China'].plot()
```

```python
corona_dataset2.loc['China'].diff().plot()
```

```python
corona_dataset2.loc['China'].diff().max()
```

```python
max_infection_rate = []
countries = corona_dataset2.index

for country in countries :
    max_infection_rate.append(corona_dataset2.loc[country].diff().max())
max_infection_rate
```

```python
corona_dataset2['max infection rate'] = max_infection_rate
```

```python
corona_dataset2.head()
```

```python
covid19_df = pd.DataFrame(corona_dataset2.loc[:,'max infection rate'])
```
